```python
# import numpy as np
vg=pd.read_csv('../lesson6/vgsales.csv')

cols=['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']
for col in cols:
    mean=vg[col].mean()
    std=vg[col].std()
    vg[col+'_Znorm']=vg[col].apply(lambda x:(x-mean)/std)

for col in cols:
    min=vg[col].min()
    max=vg[col].max()
    vg[col+'_Norm']=vg[col].apply(lambda x:(x-min)/(max-min))
vg[:20]
```
