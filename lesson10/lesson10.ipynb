### Data visualization
- Major chart types
- Basic visualization principles
- Visualization with Pandas
- Visualization with Matplotlib
- Visualization with Seaborn



```python
import pandas as pd

vg=pd.read_csv('../lesson6/vgsales.csv')
vg2=vg[['Year','Global_Sales']].groupby('Year').sum().reset_index()

vg2.plot(kind='line',x='Year')
```

```python
vg3=vg.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).reset_index()
vg3.head(10).plot(kind='bar')
```

```python
vg4=vg[['Year','Global_Sales','NA_Sales']].groupby('Year').sum().reset_index()
vg4.plot(kind='scatter',x='NA_Sales',y='Global_Sales')
```

```python
import matplotlib.pyplot as plt

vg2=vg[['Year','Global_Sales']].groupby('Year').sum().reset_index()
plt.plot(vg2['Year'],vg2['Global_Sales'])
plt.xlabel('Year')
plt.ylabel('Million')
plt.title('Global sales across years')
plt.grid(True)
plt.savefig('matplot.jpg')
# plt.show()
```

```python
vg4=vg[['Year','Global_Sales','EU_Sales','JP_Sales','NA_Sales','Other_Sales']].groupby('Year').sum().reset_index()
plt.plot(vg4.Year,vg4.Global_Sales,vg4.Year,vg4.EU_Sales,vg4.Year,vg4.JP_Sales,vg4.Year,vg4.NA_Sales,vg4.Year,vg4.Other_Sales)
# plt.plot(vg4['Year'],vg4['Global_Sales'],label='Global_Sales',color='blue')
# plt.plot(vg4['Year'],vg4['EU_Sales'],label='EU_Sales',color='yellow')
# plt.plot(vg4['Year'],vg4['JP_Sales'],label='JP_Sales',color='orange')
# plt.plot(vg4['Year'],vg4['NA_Sales'],label='NA_Sales',color='red')
# plt.plot(vg4['Year'],vg4['Other_Sales'],label='Other_Sales',color='black')
plt.xlabel('Year')
plt.ylabel('Million')
plt.title('Sales across years')
plt.grid(True)
plt.legend(['Global','EU','JP','NA','Other'])
plt.show()
```

```python
plt.rcParams['font.family']=['SimHei']
# plt.rcParams['axes.unicode_minus']=True

plt.plot([1,2,3,4,5],[2,3,4,5,6])
plt.xlabel('x')
plt.ylabel('y')
plt.title('x=y+2')
plt.grid(True)
plt.show()
```
